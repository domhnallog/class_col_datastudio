## Fiscally Standardized Cities (FiSC) data

```python
columns = ['year',
           'city_name',
           'id_city',
           'city_population',
           'rev_total',
           'rev_general',
           'chg_educ',
           'chg_nat_resources',
           'chg_parks_rec',
           'education',
           'libraries',
           'natural_resources',
           'parks_recreation',
           'capital_education',
           'capital_nat_res',
           'capital_parks_rec',
           'relationship_city_school']
fisc_df = pd.read_excel('sources/fisc_full_dataset_2016_update.xlsx', 'Dataset', usecols=columns)
```

```python
fisc_df = fisc_df[fisc_df.year == 2016].sort_values(by=['city_population'], ascending=True).reset_index(drop=True)
fisc_df = fisc_df[~fisc_df.city_name.str.contains('Average for Cities')]
fisc_df['state_name'] = fisc_df.city_name.str.extract(r'([A-Z]{2}):', expand=True)
fisc_df['city_name'] = fisc_df.city_name.str.extract(r': (.*)', expand=True)
```
